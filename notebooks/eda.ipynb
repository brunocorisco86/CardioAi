# Análise Exploratória dos Dados de Doenças Cardíacas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carregando os dados

```python
df = pd.read_csv('../data/raw/previsões de doenças cardíacas.csv')
df.head()
```

## Informações básicas do dataset

```python
df.info()
```

## Resumo estatístico

```python
df.describe()
```

## Verificando valores ausentes

```python
df.isnull().sum()
```

## Visualização de Dados

### Distribuição da variável alvo (target)

```python
sns.countplot(x='target', data=df)
plt.title('Distribuição da Variável Alvo')
plt.show()
```

### Matriz de Correlação

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()
```

### Distribuição de Idade

```python
sns.histplot(df['age'], kde=True)
plt.title('Distribuição de Idade')
plt.show()
```

### Distribuição por Sexo

```python
sns.countplot(x='sex', data=df)
plt.title('Distribuição por Sexo')
plt.xticks([0, 1], ['Feminino', 'Masculino'])
plt.show()
```

### Tipo de Dor no Peito

```python
sns.countplot(x='cp', data=df)
plt.title('Tipo de Dor no Peito')
plt.show()
```

### Relação entre Sexo e a Variável Alvo

```python
sns.countplot(x='sex', hue='target', data=df)
plt.title('Relação entre Sexo e a Variável Alvo')
plt.xticks([0, 1], ['Feminino', 'Masculino'])
plt.show()
```

### Relação entre Tipo de Dor no Peito e a Variável Alvo

```python
sns.countplot(x='cp', hue='target', data=df)
plt.title('Relação entre Tipo de Dor no Peito e a Variável Alvo')
plt.show()
```
